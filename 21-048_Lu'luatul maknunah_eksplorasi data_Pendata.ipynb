Menampilkan data dari database postgres menggunakan python

```python
#koneksi dengan database postgres dan menmpilkan data
from sqlalchemy import create_engine
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2 as ps
engine = create_engine('postgresql://postgres:luna@localhost:5432/Luluatulm')
sql = """select * from public.eda_iris"""
df = sqlio.read_sql_query(sql,engine)
print('Data keseluruhan')
df
```

```python
#menmapilkan data berdasarkan jenis species
sqls = """select * from public.eda_iris where species='Iris-setosa'"""
dfs = sqlio.read_sql_query(sqls,engine)
print('Data SPECIES SETOSA')
dfs
```

```python
sqlv = """select * from public.eda_iris where species='Iris-versicolor'"""
dfv = sqlio.read_sql_query(sqlv,engine)
print('Data SPECIES Versi color')
dfv
```

```python
sqlvir = """select * from public.eda_iris where species='Iris-virginica'"""
dfvir = sqlio.read_sql_query(sqlvir,engine)
print('Data SPECIES Versi color')
dfvir
```

Iris dataset terdiri dari 3 spesier yaitu iris setosa, iris versicolor, iris virginica dimana setiap spesiesnya didapatkan 50 sampel dengan kriteria dari masing-masing sampel antara lain panjang,lebar sepal, dan kelopak dalam sentimeter.
iris data memiliki kolom sepal length untuk panjang sepal, sepal width untuk lebar sepal, petal length untuk panjang kelopak, petal width untuk lebar kelopak.
